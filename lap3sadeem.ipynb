```python
#Q5
print("1- Addition")
print("2- Subtraction")
print("3- Multiplication")
print("4- Division")
choice = int(input("Enter Choice: "))
num1 = float(input("Enter First Number: "))
num2 = float(input("Enter Second Number: "))
match choice:
    case 1:
        result = num1 + num2
        print(num1, "+", num2, "=", result)
    case 2:
        result = num1 - num2
        print(num1, "-", num2, "=", result)
    case 3:
        result = num1 * num2
        print(num1, "*", num2, "=", result)
    case 4:
        if num2 != 0:
            result = num1 / num2
            print(num1, "/", num2, "=", result)
        else:
            print("Error: Division by zero")
    case _:
        print("Incorrect choice")
```

```python
#6
score = float(input("Enter student score: "))
if score >= 50:
    print("Passed")
else:
    print("Failed")
if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
elif score >= 50:
    grade = "E"
else:
    grade = "F"
print("Grade:", grade)
match grade:
    case "A":
        print("Excellent")
    case "B":
        print("Very Good")
    case "C":
        print("Good")
    case "D":
        print("Needs Improvement")
    case "E":
        print("Barely Passed")
    case "F":
        print("Failed")
    case _:
        print("Invalid Grade")
```
